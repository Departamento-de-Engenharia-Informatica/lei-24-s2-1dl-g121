# (1) Introduction
This scenario explores the need for a park management company to predict the average monthly water consumption cost of a new park. They possess data on the daily water consumption of existing parks and their corresponding areas. We can model this relationship using linear regression, where:

* **Dependent Variable (y):** Average monthly water consumption cost (€) for each park.
* **Independent Variable (X):** Park area (hectares).

The following equation represents the general form of a linear regression model:

$$y = \beta_0 + \beta_1 X + \varepsilon$$

-  $\beta_0$  represents the y-intercept, the predicted average monthly cost when the park area is zero (which is unlikely but helps interpret the model).
-  $\beta_1$  represents the slope, indicating the change in average monthly cost associated with a one-hectare increase in park area.
-  $\varepsilon$  represents the error term, capturing the random variation not explained by the model.

This code utilizes linear regression to analyze the data and predict the average monthly water cost for a new 55-hectare park.

# (2) Code and results

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Load the data
water_consumption = pd.read_csv("water_consumption_updated.csv", delimiter=';', decimal=',')
area_data = pd.read_csv("Area.csv", delimiter=';', decimal=',')

# Strip any leading/trailing spaces from column names (if necessary)
water_consumption.columns = water_consumption.columns.str.strip()
area_data.columns = area_data.columns.str.strip()

# Function to calculate monthly water consumption cost
def calculate_monthly_cost(consumption):
    if consumption <= 1000:
        return consumption * 0.7
    else:
        base_cost = 1000 * 0.7
        additional_cost = (consumption - 1000) * 0.7 * 1.15
        return base_cost + additional_cost

# Aggregate daily consumption to monthly consumption
water_consumption['Date'] = pd.to_datetime(water_consumption[['Year', 'Month', 'Day']])
monthly_consumption = water_consumption.groupby(['Park', 'Year', 'Month'])['Consumption'].sum().reset_index()

# Calculate the cost
monthly_consumption['Cost'] = monthly_consumption['Consumption'].apply(calculate_monthly_cost)

# Calculate the average monthly cost per park
average_monthly_cost = monthly_consumption.groupby('Park')['Cost'].mean().reset_index()
average_monthly_cost.columns = ['Park', 'Average Monthly Cost']

# Merge with area data
park_data = pd.merge(average_monthly_cost, area_data, on='Park')

display(average_monthly_cost)

# Prepare data for linear regression
X = park_data['Area'].values.reshape(-1, 1)  # Independent variable (Area)
y = park_data['Average Monthly Cost'].values  # Dependent variable (Cost)

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict the cost for the new 55-hectare park
new_park_area = np.array([[55]])
predicted_cost = model.predict(new_park_area)

# Calculate the confidence interval
n = len(X)
dof = n - 2
t_stat = t.ppf(0.975, dof)
residuals = y - model.predict(X)
std_error = np.sqrt(np.sum(residuals ** 2) / dof)
confidence_interval = t_stat * std_error * np.sqrt(1 + 1 / n + (new_park_area - np.mean(X)) ** 2 / np.sum((X - np.mean(X)) ** 2))

print(f"\nPredicted average monthly cost for the 55-hectare park: {predicted_cost[0]:.2f} €")
print(f"95% CI: ({predicted_cost[0] - float(confidence_interval):.2f} €, {predicted_cost[0] + float(confidence_interval):.2f} €)")

# Visualize the data points and the regression line
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression line')
plt.scatter(new_park_area, predicted_cost, color='green', marker='x', s=100, label='Prediction for 55-hectare park')

plt.xlabel('Park Area (hectares)')
plt.ylabel('Average Monthly Cost (€)')
plt.title('Linear Regression: Park Area vs. Average Monthly Water Cost')
plt.legend()
plt.grid(True)
plt.show()
```

# (3) Analysis and interpretation of the results
The code performs a linear regression analysis to estimate the relationship between park area and water consumption cost. However, a more comprehensive analysis can be conducted to assess the model's validity and interpret the results:

* **Model Evaluation:**
    - Calculate the R-squared value to quantify how well the linear model fits the data. A higher R-squared (closer to 1) indicates a better fit.
    - Analyze residual plots to check for normality of residuals and identify potential outliers that might significantly impact the model.

* **Interpretation of Coefficients:**
    - Analyze the estimated values of $\beta_0$ and $\beta_1$ obtained from the linear regression model.
    - The intercept, $\beta_0$, reflects the predicted average monthly cost when the park area is zero (likely not applicable but provides a reference point).
    - The slope, $\beta_1$, indicates the change in average monthly cost associated with a one-hectare increase in park area. A positive slope suggests that larger parks incur higher average monthly water costs.

* **Prediction Uncertainty:**
    - The confidence interval around the predicted cost for the new park reflects the model's uncertainty. A wider confidence interval indicates higher prediction uncertainty.

By performing this additional analysis, we can gain a deeper understanding of the following:

* How well the linear model captures the relationship between park area and water consumption cost.
* The impact of park size on the average monthly water cost.
* The level of confidence associated with the predicted cost for the new park.

This will enable the park management company to make more informed decisions regarding water consumption management and budgeting for the new park.
